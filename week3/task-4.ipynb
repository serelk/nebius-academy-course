```python
retriever = loaded_vector_store.as_retriever(
    search_type="similarity",
    search_kwargs={
        "k": 8,
        "fetch_k": 50,
        # "lambda_mult": 0.5,
    },
)

```

```python
def rewrite_query(question: str) -> str:
      response = client.chat.completions.create(
          model="meta-llama/Llama-3.3-70B-Instruct",
          messages=[
              {"role": "system", "content": (
                  "You are a search query optimizer for financial document retrieval. "
                  "Given a question, rewrite it as a search query that would match "
                  "the actual text in SEC filings (10-K, 10-Q, annual reports). "
                  "Replace high-level financial concepts with the specific line items, "
                  "table headers, and section titles that would appear in the filing. "
                  "Output only the rewritten query, nothing else."
              )},
              {"role": "user", "content": question}
          ],
          temperature=0
      )
      return response.choices[0].message.content
```

```python
def hyde_retrieve(question: str, k: int = 10):
      # Step 1: Generate hypothetical document
      response = client.chat.completions.create(
          model="meta-llama/Llama-3.3-70B-Instruct",
          messages=[
              {"role": "system", "content": (
                  "You are a financial document generator. Given a question, "
                  "write a short passage (2-3 sentences) that would appear in "
                  "an SEC filing (10-K, 10-Q) and answer this question. "
                  "Use realistic financial statement formatting, line item names, "
                  "and plausible numbers. Output only the passage."
              )},
              {"role": "user", "content": question}
          ],
          temperature=0
      )
      hypothetical_doc = response.choices[0].message.content

      # Step 2: Search using the hypothetical doc as query
      results = loaded_vector_store.similarity_search(hypothetical_doc, k=k)
      return results, hypothetical_doc

  # Usage:

  # retrieved_docs, hypo = hyde_retrieve(question_1)
  # print("Hypothetical doc:", hypo)
  # print()
  # for doc in retrieved_docs:
  #     print(doc.metadata["page"])
```
